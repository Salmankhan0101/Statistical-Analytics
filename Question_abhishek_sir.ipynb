```python
# Q1.write a python program to find the mean of a list of number.
```

```python
data=[2,5,7,9,5]
s=0
c=0
for x in data:
  s=s+x
  c=c+1
print(s/c)
```

```python
# Q2.write a python program to find the median of a list of numbers without using any library.
```

```python
data=[2,5,7,8,9]
data.sort()
N=len(data)
if N%2==0:
  median=(data[N//2]+data[N//2-1])/2
else:
  median=data[N//2]
print(median)
```
